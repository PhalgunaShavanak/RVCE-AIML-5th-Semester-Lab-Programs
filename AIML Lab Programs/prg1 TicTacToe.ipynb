```python
def print_brd(brd):
    for row in brd:
        print("   |   ".join(row))
        print("-----------")

def check_win(brd, player):
    for i in range(3):
        if all(brd[i][j] == player for j in range(3)) or all(brd[j][i] == player for j in range(3)):
            return True
        if all(brd[i][i] == player for i in range(3)) or all(brd[i][2-i] == player for i in range(3)):
            return True
    return False

def is_full(brd):
    return all(brd[i][j] != '' for i in range(3) for j in range(3))

def dfs(brd, player):
    if check_win(brd, 'X'):
        return 1
    if check_win(brd, 'O'):
        return -1
    elif is_full(brd):
        return 0
    scores = []
    for i in range(3):
        for j in range(3):
            if brd[i][j] == '':
                brd[i][j] = player
                if player == 'X':
                    scores.append(dfs(brd, 'O'))
                else:
                    scores.append(dfs(brd, 'X'))
                brd[i][j] = ''
    if player == 'X':
        return max(scores)
    else:
        return min(scores)

def best_move(brd):
    best_score = float('-inf')
    best_move = (-1, -1)
    for i in range(3):
        for j in range(3):
            if brd[i][j] == '':
                brd[i][j] = 'X'
                score = dfs(brd, 'O')
                brd[i][j] = ''
                if score > best_score:
                    best_score = score
                    best_move = (i, j)
    return best_move

def start_game():
    brd = [['' for _ in range(3)] for _ in range(3)]
    curr_player = 'X'
    while True:
        print_brd(brd)
        if curr_player == 'X':
            row, col = best_move(brd)
            brd[row][col] = 'X'
        else:
            row = int(input("Enter row (0/1/2): "))
            col = int(input("Enter col (0/1/2): "))
            if 0 <= row <= 2 and 0 <= col <= 2 and brd[row][col] == '':
                brd[row][col] = 'O'
            else:
                print("Invalid input, try again.")

        if check_win(brd, curr_player):
            print_brd(brd)
            print(f"Player {curr_player} wins!")
            break
        elif is_full(brd):
            print_brd(brd)
            print("Tie!")
            break

        curr_player = 'X' if curr_player == 'O' else 'O'

start_game()
```
